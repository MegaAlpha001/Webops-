```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**5 - 13*x**3 - x**2 + 10*x + 170

x = np.linspace(-5, 5, 400)
y = f(x)

plt.figure(figsize=(8, 8))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x^5 - 13x^3 - x^2 + 10x + 170')
plt.grid(True)
plt.show()
```
